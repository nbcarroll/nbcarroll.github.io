### Aircraft Registration Master file

```python
# Replace 'your_file.txt' with the path to your text file
df_aircraft_registration_2015 = pd.read_csv('MASTER.txt', sep=',')
```

```python
df_aircraft_registration_2015.info()
```

```python
df_aircraft_registration_2015.head()
```

We can see multiple columns that have date/time in a format that is not datetime, so we'll correct this

```python
df_aircraft_registration_2015['LastActivity Date'] = pd.to_datetime(df_aircraft_registration_2015['LastActivity Date'], errors='coerce')
df_aircraft_registration_2015['CertIssueDate'] = pd.to_datetime(df_aircraft_registration_2015['CertIssueDate'], errors='coerce')
df_aircraft_registration_2015['Airworthiness Date'] = pd.to_datetime(df_aircraft_registration_2015['Airworthiness Date'], errors='coerce')
df_aircraft_registration_2015['Expiration Date'] = pd.to_datetime(df_aircraft_registration_2015['Expiration Date'], errors='coerce')
```

Additional Data Diving

```python
# What is the count of aircraft that expire in 2015 or later?
df_aircraft_registration_2015['Expiration Date'].dt.year.value_counts().sort_index()

# Show only aircraft rows that expire before 2011
df_aircraft_registration_2015[df_aircraft_registration_2015['Expiration Date'].dt.year < 2011]
```
